```python
import requests
import json
import os
from pathlib import Path
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from dotenv import load_dotenv
load_dotenv()
```

```python
api_key = os.getenv("rapidapi_key")
url = "https://rapidapi.p.rapidapi.com/api/search/NewsSearchAPI"
query = "ethereum"
page_number = 1
page_size = 100
auto_correct = True
safe_search = False
with_thumbnails = False
from_published_date = "2021-01-01"
to_published_date = "2021-01-24"
querystring = {"q": query,
               "pageNumber": page_number,
               "pageSize": page_size,
               "autoCorrect": auto_correct,
               "safeSearch": safe_search,
               "withThumbnails": with_thumbnails,
               "fromPublishedDate": from_published_date,
               "toPublishedDate": to_published_date}
headers = {
    'x-rapidapi-key': api_key,
    'x-rapidapi-host': "contextualwebsearch-websearch-v1.p.rapidapi.com"
    }

response = requests.request("GET", url, headers=headers, params=querystring)
analyzer = SentimentIntensityAnalyzer()
eth_sentiments = []
eth_headlines = response.json()['value']

# for article in eth
eth_headlines
eth_sentiments = []
for article in eth_headlines:
    try:
        text = article["body"]
        date = article["datePublished"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        eth_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
        })
    except AttributeError:
        pass
    

eth_df_1 = pd.DataFrame(eth_sentiments)
eth_df_1
```
