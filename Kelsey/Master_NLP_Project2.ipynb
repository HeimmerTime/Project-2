```python
# Initial imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from pathlib import Path

%matplotlib inline
```

```python
from numpy.random import seed
seed(1)

from tensorflow import random
random.set_seed(2)
```

```python
# Import the dataset
file_path = Path("../Kelsey/ETH_Bull.csv")
eth_bull_news_df = pd.read_csv(file_path, index_col = 'date', parse_dates=True, infer_datetime_format=True)
eth_bull_news_df = eth_bull_news_df.sort_index()
eth_bull_news_df
```

```python
eth_bull_news_df.insert(5, 'market', 'Bull')
eth_bull_news_df.head()
```

```python
# Import the dataset
file_path = Path("../Kelsey/ETH_Bear.csv")
eth_bear_news_df = pd.read_csv(file_path, index_col = 'date', parse_dates=True, infer_datetime_format=True)
eth_bear_news_df = eth_bear_news_df.sort_index()
eth_bear_news_df.head()
```

```python
eth_bear_news_df.insert(5, 'market', 'Bear')
eth_bear_news_df.head()
```

```python
eth_news_combined = pd.concat([eth_bull_news_df, eth_bear_news_df], axis="rows", join="inner")
eth_news_combined
```

```python
# Drop first row
# by selecting all rows from first row onwards
eth_news_combined.drop('0001-01-01', inplace=True)
eth_news_combined
```

```python
import datetime as dt
```

```python
from sklearn.preprocessing import LabelEncoder
# Create the LabelEncoder instance
le = LabelEncoder()

# adding column
le.fit(eth_news_combined["market"])
eth_news_combined["market"] = le.transform(eth_news_combined["market"])
eth_news_combined.head()
```

```python
eth_news_combined.index.values
```

```python
eth_news_combined.index = pd.to_datetime(eth_news_combined.index)
```

```python
X = eth_news_combined["text"].values
y = eth_news_combined["market"].values
```

```python
# Import the Tokenizer method from Keras
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
# Create an instance of the Tokenizer and fit it with the X text data
tokenizer = Tokenizer(lower=True)
tokenizer.fit_on_texts(X)
```

```python
# Print the first five elements of the encoded vocabulary
for token in list(tokenizer.word_index)[:5]:
    print(f"word: '{token}', token: {tokenizer.word_index[token]}")
```

```python
# Transform the text data to numerical sequences
X_seq = tokenizer.texts_to_sequences(X)
```

```python
# Contrast a sample numerical sequence with its text version
print("**Text comment**")
print({X[0]})
```

```python
print("**Numerical sequence representation**")
print(X_seq[0])
```

```python
# Import the pad_sequences method from Keras
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# Set the pad size
max_words = 160

# Pad the sequences using the pad_sequences() method
X_pad = pad_sequences(X_seq, maxlen=max_words, padding="post")
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Creating training, validation, and testing sets using the encoded data
X_train_rnn, X_test_rnn, y_train_rnn, y_test_rnn = train_test_split(X_pad, y)

#X_train_rnn, X_val_rnn, y_train_rnn, y_val_rnn = train_test_split(X_train_rnn, y_train_rnn)
```

```python
y_test_rnn
```

```python
# Import Keras modules for model creation
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
# Model set-up
vocabulary_size = len(tokenizer.word_counts.keys()) + 1
embedding_size = 64
```

```python
# Define the LSTM RNN model
model = Sequential()

# Layer 1
model.add(Embedding(vocabulary_size, embedding_size, input_length=max_words))

# Layer 2
model.add(LSTM(units=320))

# Output layer
model.add(Dense(units=1, activation="sigmoid"))
```

```python
# Compile the model
model.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=[
        "accuracy",
        tf.keras.metrics.TruePositives(name="tp"),
        tf.keras.metrics.TrueNegatives(name="tn"),
        tf.keras.metrics.FalsePositives(name="fp"),
        tf.keras.metrics.FalseNegatives(name="fn"),
        tf.keras.metrics.Precision(name="precision"),
        tf.keras.metrics.Recall(name="recall"),
        tf.keras.metrics.AUC(name="auc"),
    ],
)
```

```python
# Show model summary
model.summary()
```

```python
# Training the model
batch_size = 200
epochs = 40
model.fit(
    X_train_rnn,
    y_train_rnn,
    epochs=epochs,
    batch_size=batch_size,
    verbose=1,
)
```

```python
# Predict classes using the testing data
y_rnn_pred = model.predict_classes(X_test_rnn, batch_size=200)
```

```python
# Accuracy
from sklearn.metrics import accuracy_score

#print("Vader Accuracy: %.2f" % (accuracy_score(y_test, y_vader_pred)))
#print("RNN LSTM Accuracy %.2f" % (accuracy_score(y_test_rnn, y_rnn_pred)))
```

```python
# Import the confusion_matrix method from sklearn
from sklearn.metrics import confusion_matrix
```

```python
# Confusion matrtix metrics from the RNN LSTM model
tn_rnn, fp_rnn, fn_rnn, tp_rnn = confusion_matrix(y_test_rnn, y_rnn_pred).ravel()

# Dataframe to display confusion matrix from the RNN LSTM model
cm_rnn_df = pd.DataFrame(
    {
        "Positive(1)": [f"TP={tp_rnn}", f"FP={fp_rnn}"],
        "Negative(0)": [f"FN={fn_rnn}", f"TN={tn_rnn}"],
    },
    index=["Positive(1)", "Negative(0)"],
)
cm_rnn_df.index.name = "Actual"
cm_rnn_df.columns.name = "Predicted"
print("Confusion Matrix from the RNN LSTM Model")
display(cm_rnn_df)
```

```python
# Import the classification_report method from sklearn
from sklearn.metrics import classification_report
```

```python
# Display classification report for the RNN LSTM Model
print("Classification Report for the RNN LSTM Model")
print(classification_report(y_rnn_pred, y_test_rnn))
```
