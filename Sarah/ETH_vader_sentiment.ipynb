```python
# Initial imports
import os
from pathlib import Path
import pandas as pd
from newsapi import NewsApiClient
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from dotenv import load_dotenv
load_dotenv()
```

```python
# Download/Update the VADER Lexicon
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

```

```python
# Load ETH from CSV file
file_path = Path("Data_Ethereum2_EthTrader.csv")
eth_df = pd.read_csv(file_path, encoding="utf-8-sig", parse_dates=True, infer_datetime_format=True)
eth_df.head()
```

```python
## Create the Ethereum sentiment scores DataFrame
Ethereum_sentiments = []

for title in eth_df:
    try:
        text = title
        #date = title["date"]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        Ethereum_sentiments.append({
            
            
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu,
            "text": text,
            #"date": date
            
        })
        
    except AttributeError:
        pass
    

eth_sentiment_df = pd.DataFrame(Ethereum_sentiments)
```
